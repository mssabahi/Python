## 2- Plotting Differences

### Plot the difference between the average of the first dataset and the average of the second dataset, i.e., the difference between the leftmost plot of the first two figures.

```python
import glob
filenames = sorted(glob.glob('data/inflammation*.csv'))


data_0 = np.loadtxt(filenames[0], delimiter=',')
data_1 = np.loadtxt(filenames[1], delimiter=',')

data_0_mean = np.mean(data_0, axis = 0)
data_1_mean = np.mean(data_1, axis = 0)

data_diff = data_0_mean-data_1_mean
```

```python
filenames
```

```python
fig = plt.figure(figsize = (10,4))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(data_diff)
axes.set_xlim(0,data_diff.shape[0])
axes.set_ylim(min(data_diff),max(data_diff))
axes.set_xlabel('Measurment')
axes.set_ylabel('difference of averages')
```

## 6- Palindrome : A palindrome is a word that is spelled the same backward and forward, like “noon” and “redivider”. Recursively, a word is a palindrome if the first and last letters are the same and the middle is a palindrome. 

### -Given a string, find the longest palindrome for that string.

```python
text = input('Enter your word: ')
#text = 'redivider'
my_str = text
f_text = []
j=0

while j<= len(text):
    for i in range(3,len(my_str)+1):
        my_strn = my_str[0:i]
        if (str(my_strn) == str(my_strn)[::-1]):
            f_text.append(my_strn) 
    j =j+1
    my_str = text[j:]

for i in range(0,len(f_text)):
    print(f'The palindrome "{f_text[i]}" in the input word has the length of {len(f_text[i])}.')
print(f'The longest polindrom is "{max(f_text,key=len)}" with the length of {len(max((f_text),key=len))} ')
```
