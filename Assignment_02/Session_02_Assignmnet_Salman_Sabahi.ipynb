## Generate Composite Statistics

### Use each of the files once to generate a dataset containing values averaged over all patients. Then use pyplot to generate average, max, and min for all patients.

```python
import glob
file_names = sorted(glob.glob('data/inflammation*.csv'))
average = []
maximum = []
minimum = []

data = []
for i in range(0,len(file_names)):
    data.append(np.loadtxt(file_names[i], delimiter=','))

for i in range(0,len(data)):
    average.append(np.mean(data[i], axis = 0))
    maximum.append(np.max(data[i], axis = 0))
    minimum.append(np.min(data[i], axis = 0))
    
    
    
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (10,4), dpi = 100)
    
    axes[0].plot(average[i])
    axes[1].plot(maximum[i])
    axes[2].plot(minimum[i])
    
    axes[0].set_xlim(0,data[i].shape[1])
    axes[0].set_ylim(0,max(average[i]))

    axes[1].set_xlim(0,data[i].shape[1])
    axes[1].set_ylim(0,max(maximum[i]))

    axes[2].set_xlim(0,data[i].shape[1])
    axes[2].set_ylim(0,max(minimum[i]))
    
    
    axes[0].set_ylabel('Mean ')
    axes[1].set_ylabel('Max ')
    axes[2].set_ylabel('Min ')

    axes[0].set_xlabel('Measurment ')
    axes[1].set_xlabel('Measurment ')
    axes[2].set_xlabel('Measurment ')
    
    axes[0].set_title(f' {file_names[i]}', fontsize = 10)
    axes[1].set_title(f' {file_names[i]}', fontsize = 10)
    axes[2].set_title(f' {file_names[i]}', fontsize = 10)
    
    fig.tight_layout()


```

## The greatest common divisor (GCD) of a and b is the largest number that divides both of them with no remainder. 
### One way to find the GCD of two numbers is based on the observation that if r is the remainder when a is divided by b, then gcd(a, b) = gcd(b, r). As a base case, we can use gcd(a, 0) = a.
### Write a function called gcd that takes parameters a and b and returns their greatest common divisor


```python
def GCD(num_1,num_2):
    assert num_1 >= num_2, "The first number has to be greater than the second one"
    rem = num_1 % num_2
    while (rem != 0 & rem != 1) :
        num_1 = num_2
        num_2 = rem
        rem = num_1 % num_2
    if rem == 0:
        print(f'The Greatest Common Divisor is {num_2}')
    else: 
        print(f'The Greatest Common Divisor is 1')       
```

```python
# with recursion

import turtle
myTurtle = turtle.Turtle()
myWin = turtle.Screen()

myTurtle.up()
myTurtle.setposition(-100,100)
myTurtle.down()

def snow_flake(vec,len_line,main_loop=3,num=0):
    '''
    This function draws a snowflake. To do that you need to provide two components as input.
    vec: It is a vector that changes the angles inside the snowflake. 
    len_line: It is the length of the snowflake's lines. 
    main_loop: It is the number of main lines which compose a snowflake
    (This number is equal to 3 because the simplest form of a snowflake is a triangle). This values determines when the loop must end.
    num: This component is a numerator which must be zero.
    '''
    for s_item in range(0,len(vec)-1):
        myTurtle.forward(len_line)
        myTurtle.left(vec[s_item])
    myTurtle.forward(len_line)
    
    if num <= s_item:
        myTurtle.left(vec[num])
        snow_flake(vec,len_line,main_loop,num+1)
        
    elif num > s_item & main_loop > 0:
        num = 0
        myTurtle.left(vec[-1])
        snow_flake(vec,len_line,main_loop-1,num)   

    myWin.exitonclick()

snow_flake([60,-120,60,-120],20)
```
